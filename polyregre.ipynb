```python
import  pandas as pd
import matplotlib.pyplot as plt
```

```python
# data loading
df=pd.read_csv('http://13.234.66.67/summer19/datasets/salary.csv')
```

```python
exp=df.iloc[:,0:1].values # exp
```

```python
# sal
sal=df.iloc[:,1:].values
```

```python
# call linear model just for compare
from sklearn.linear_model import LinearRegression
regr1=LinearRegression()
```

```python
# to transfer data we can call poly transform method
from sklearn.preprocessing import PolynomialFeatures
```

```python
# experience transform
poly=PolynomialFeatures(degree=15)
```

```python
# exp fit
newexp=poly.fit_transform(exp)
```

```python
# apply linear regression
trained1=regr1.fit(exp,sal)
```

```python
# now we can visualize this experience and saary data
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(exp,sal,label="actual data as per exp")
plt.plot(exp,trained1.predict(exp),label="predicted salary")
plt.legend()
plt.show()
```

```python
# apply new exp data in mmodel
regr2=LinearRegression()
```

```python
#  new exp data only
trained2=regr2.fit(newexp,sal)
```

```python
# now we can visualize this experience and saary data
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(exp,sal,label="actual data as per exp")
plt.plot(exp,trained2.predict(newexp),label="new model")
plt.legend()
plt.show()
```
