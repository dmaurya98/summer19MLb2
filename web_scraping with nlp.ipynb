```python
import requests
from bs4 import BeautifulSoup
import html5lib
```

```python
webdata=requests.get('https://php.net')
```

```python
webdata
```

```python
htmldata=webdata.text
```

```python
htmldata
```

```python
soup=BeautifulSoup(htmldata,"html5lib")
```

```python
clean_data=soup.get_text()
```

```python
with open("mywebdata.txt",'w')  as f:
  f.write(clean_data)
```

```python
# 
f= open("mywebdata.txt",'r')
mydata=f.read()
f.close()
```

```python
newdata=[i for i in mydata.split()]
```

```python
import matplotlib.pyplot as plt
import nltk
```

```python
[i for i in dir(nltk) if 'Freq' in i] # this will plot data
```

```python
nlpdata=nltk.FreqDist(newdata)
```

```python
nlpdata.plot(20)
```
